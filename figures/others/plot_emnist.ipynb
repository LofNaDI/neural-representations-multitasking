```python
import os
import pickle

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import torch.nn as nn
import torch.optim
from tqdm.auto import tqdm

import multitask.dataset as dataset
```

```python
tasks_datasets = dataset.get_tasks_dict(['vowel'], root='../../data')
```

```python
idxs_numbers = {}
labels = tasks_datasets['vowel'].letters

for letter in range(26):
    labels_number = np.where(labels == letter)[0]
    idxs_numbers[letter] = labels_number.tolist()
```

```python
idxs_plot = 0

data = tasks_datasets['vowel'].data

for letter in range(26):
    fig = plt.figure(figsize=(4, 4))
    plt.imshow(data[idxs_numbers[letter][idxs_plot]].reshape(28, 28),
               cmap='gray')
    plt.axis('off')
    fig.savefig(f'letters/letter_{chr(97 + letter)}.pdf', transparent=True)
plt.show()
```
